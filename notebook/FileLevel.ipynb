```python
import pandas as pd
import re
import numpy as np
```

```python
wordlist=[]
with open('C:/Users/avson/Desktop/research/notebook/filelevelcommit.txt', 'r') as f:
    wordlist = f.read().strip().splitlines()
len(wordlist)
```

```python
for i in range(len(wordlist)):
    wordlist[i]=wordlist[i].strip()
```

```python
wordlist[7023]
```

```python
filenames=[]
dates=[]
changes=[]
additions=[]
deletions=[]
i=0
while (i<=len(wordlist)-3):
    try:
        temp=wordlist[i+1]
        match=re.search(r'.* change[s]?:',temp)
        changes.append(temp[match.start():match.end()-8])
        match=re.search(r':.* addition[s]?',temp)
        additions.append(temp[match.start()+2:match.end()-9].strip())
        match=re.search(r'&.* deletion[s]?',temp)
        deletions.append(temp[match.start()+2:match.end()-9].strip())
        filenames.append(wordlist[i])
        dates.append(wordlist[i+2])
    except Exception as e:
        print(e)
    i+=3
```

```python
len(dates)
```

```python
finaldict={'filename':filenames,'additions':additions,'deletions':deletions,'changes':changes,'date':dates}
data=pd.DataFrame(finaldict)
```

```python
data.to_csv('fileleveldata.csv')
```
